```python
# Link to the uploaded dataset on Kaggle
link = "https://www.kaggle.com/datasets/nishthasharma01/sustainable-environment-questions-and-answers"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL of the website to scrape
url = "https://www.futurelearn.com/info/blog/50-sustainability-faqs-answered"
```

```python
# Send a GET request to the website
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:

    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all question and answer elements
    questions_and_answers = soup.find("div", class_="entry-content")

    questions, answers = [], []

    # Extract questions
    for h3 in questions_and_answers.find_all("h3"):

        question_text = h3.text.strip()
        questions.append(question_text)

        answer_parts = [] # for multi paragraph answers

        # Find all p tags between current h3 and the next h3
        for p in h3.find_next_siblings(["p", "h3"]):

            if p.name == "h3":
                break  # Stop if the next h3 is encountered

            answer_parts.append(p.text.strip())

        answer_text = " ".join(answer_parts)

        # for not adding empty paragraphs
        if answer_text:
            answers.append(answer_text)

else:
    print(f"Error: Could not access the website. Status code: {response.status_code}")
```

```python
import pandas as pd
```

```python
questions_and_answers = {"questions": questions, "answers": answers[: 50]}
dataset = pd.DataFrame(questions_and_answers)
dataset
```

```python
dataset.to_csv("sustainable_practices_qna.csv")
```
